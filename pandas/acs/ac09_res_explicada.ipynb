```python
df["Cidade"] = df["Territorialidade"].str.split("*", expand = True)[2]
df["Estado"] = df["Territorialidade"].str.split("*", expand = True)[1]
df["Região"] = df["Territorialidade"].str.split("*", expand = True)[0]

# CO*GO*Abadia de Goiás * -> ["CO", "GO", "Abadia de Goiás "]

df.head()
```

```python
df.columns
```

```python
df2 = df.filter(items=['Região', 'Estado', 'Cidade', 'Posicao IDHM 2010', 'IDHM 2010',
       'Posicao IDHM Renda 2010', 'IDHM Renda 2010',
       'Posicao IDHM Educacao 2010', 'IDHM Educacao 2010',
       'Posicao IDHM Longevidade 2010', 'IDHM Longevidade 2010',
       'Posicao IDHM 2000', 'IDHM 2000', 'Posicao IDHM Renda 2000',
       'IDHM Renda 2000', 'Posicao IDHM Educacao 2000', 'IDHM Educacao 2000',
       'Posicao IDHM Longevidade 2000', 'IDHM Longevidade 2000'])
```

```python
df2.head()
```

**Q01)** Qual foi a cidade do Ceará que teve o maior crescimento do “IDHM Educação” entre os anos 2000 e 2010?

```python
df2["Taxa Educacao"] = (df["IDHM Educacao 2010"] - df["IDHM Educacao 2000"]) / df["IDHM Educacao 2000"]
```

```python
df2.head()
```

```python
q1 = df2["Cidade"].loc[df2["Taxa Educacao"] == df2.where(df2["Estado"] == "CE")["Taxa Educacao"].max()]
q1
```

**Q02)** Qual foi o “IDHM Longevidade 2010” na cidade que responde corretamente o item anterior?

```python
# q2 = df2["IDHM Longevidade 2010"].loc[4313] -> Método caso o índice (ou linha) da cidade já seja conhecida

q2 = df2["IDHM Longevidade 2010"].loc[df2["Taxa Educacao"] == df2.where(df2["Estado"] == "CE")["Taxa Educacao"].max()]
q2
```

**Q03)** Qual foi a cidade de Pernambuco que teve a menor “IDHM Renda 2000”?

```python
df_PE = df2.where(df2["Estado"] == "PE")
```

```python
q3 = df_PE["Cidade"].loc[df_PE["IDHM Renda 2000"] == df_PE["IDHM Renda 2000"].min()]
q3
```

**Q04)** Qual foi o crescimento percentual do “IDHM Educação” na cidade que responde corretamente o item anterior entre os anos 2000 e 2010?

```python
q4 = df_PE["Taxa Educacao"].loc[df_PE["IDHM Renda 2000"] == df_PE["IDHM Renda 2000"].min()]
q4
```

**Q05)** Qual foi o “IDHM Longevidade 2010” da cidade da Bahia que teve o maior “IDHM Educacao 2000”?

```python
df_BA = df2.where(df2["Estado"] == "BA")
q5 = df_BA["IDHM Longevidade 2010"].loc[df_BA["IDHM Educacao 2000"] == df_BA["IDHM Educacao 2000"].max()]
q5
```
