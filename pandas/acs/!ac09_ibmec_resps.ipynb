```python
#No 2o passo, devemos fazer uma análise prévia do DF
#Apareceu algum caracter meio duvidoso? tipo &, %, $, #, @...
#Será que as colunas que deveriam ter dados numéricos estão como Float ou Int? (função INFO)

df.info()
```

```python
#Como está tudo OK, vamos para o 3o passo
#Nesse caso, vamos dividir a coluna Territorialidade em 3 novas colunas

df['Cidade']=df.Territorialidade.str.split('*',expand=True)[0]
df['UF']=df.Territorialidade.str.split('*',expand=True)[1]
df['Região']=df.Territorialidade.str.split('*',expand=True)[1]
df.head()

#Propositadamente com ERRO!!!
```

```python
#corrigindo o ERRO: atenção para os números que aparecem nos []

df.Territorialidade.str.split('*',expand=True)
```

```python
df['Cidade']=df.Territorialidade.str.split('*',expand=True)[2]
df['UF']=df.Territorialidade.str.split('*',expand=True)[1]
df['Região']=df.Territorialidade.str.split('*',expand=True)[0]
df.head()
```

### **Apenas mostrando duas formas de apagar linhas ou colunas com o DROP**

```python
df.drop('Posicao IDHM 2010', axis=1, inplace=True)
df.head()
```

```python
df.drop('Posicao IDHM 2000', axis=1)
df.head()

#se não colocar o "inplace", o df não fica alterado
```

```python
df.drop(1, inplace=True)
df.head()

#se não escrever "axis=1", ele considera "axis=0"
```

### **Voltando a solução da AC09**

```python
#Ainda precisamos criar a coluna que calcula a taxa de crescimento entre os anos 2000 e 2010

df['taxa']=(df['IDHM Educacao 2010'] - df['IDHM Educacao 2000'])/df['IDHM Educacao 2000']
df.head()
```

```python
#Agora que já temos todas as colunas, podemos passar para o 4o passo, QUE NÃO É OBRIGATÓRIO!!!
#Mas facilita a visualização e o processamento
#Vamos usar a função COLUMNS e depois a função FILTER para fazer um "Filtro de colunas"
df.columns
```

```python
df2=df.filter(items=['Região','UF','Cidade',
                     'IDHM 2000','IDHM 2010',
                     'IDHM Educacao 2000', 'IDHM Educacao 2010','taxa',
                     'IDHM Renda 2000','IDHM Renda 2010',
                     'IDHM Longevidade 2000','IDHM Longevidade 2010'])
df2.head()
```

## **Q01**

```python
#Depois de filtrado e organizado, vamos para o 5o passo
#01) Qual foi a cidade do Ceará que teve o maior crescimento do “IDHM Educação” entre os anos 2000 e 2010?

q1 = df2['Cidade'].loc[df2['taxa']
                     == df2.where(df2['UF']=='CE')
                     ['taxa'].max()]
q1
```

```python
#Apenas para visualizarmos os trechos do código
#Esse comando traz um valor, nesse caso, numérico

df2.where(df2['UF']=='CE')['taxa'].max()
```

```python
#Cuidado ao copiar algum valor numérico e não utilizar todas as casas decimais

df2['Cidade'].loc[df2['taxa'] == 2.4173228]
```

```python
#No trecho acima não foi encontrada nenhuma cidade,...
#... pois não foi informado o número inteiro, com todas as casas decimais

df2['Cidade'].loc[df2['taxa']==2.4173228346456694]
```

```python
#Outro detalhe, o código df['Cidade'], aponta para coluna "Cidade" da linha 4313

df2.loc[df2['taxa']==2.4173228346456694]
```

```python
#Para responder a 2a questão, nem precisaria saber que o nome da cidade é Salitre
#02) Qual foi o “IDHM Longevidade 2010” na cidade que responde corretamente o item anterior?

q2=df2['IDHM Longevidade 2010'].loc[df2['taxa']==df2.where(df2['UF']=='CE')['taxa'].max()]
q2
```

```python
#Repetindo o mesmo processo da Q1
#03) Qual foi a cidade de Pernambuco que teve a menor “IDHM Renda 2000”?

df2['Cidade'].loc[df2['IDHM Renda 2000']
                  == df2.where(df2['UF']=='PE')
                  ['IDHM Renda 2000'].min()]
```

```python
#Porque apareceu mais de uma cidade?
df2.loc[df2['IDHM Renda 2000']==df2.where(df2['UF']=='PE')['IDHM Renda 2000'].min()]
```

```python
#Isso acontece porque ele está localizando no DF com todas as cidades do Brasil

df2.loc[df2['IDHM Renda 2000'] == 0.366]
```

```python
#Para amenizar essa situação, vamos fazer um "filtro de linhas"

df_PE = df2.where(df2['UF']=='PE').dropna()
df_PE
```

```python
#Agora que temos um DF apenas com cidades de PE, podemos usar o mesmo código

q3=df_PE['Cidade'].loc[df_PE['IDHM Renda 2000']
                       == df2.where(df2['UF']=='PE')
                       ['IDHM Renda 2000'].min()]
q3
```

```python
#De forma semelhante, a Q4 poderia ser respondida diretamente, sem saber a resposta da Q3
#04) Qual foi o crescimento percentual do “IDHM Educação” na cidade que responde corretamente o item anterior entre os anos 2000 e 2010?

q4=df_PE['taxa'].loc[df_PE['IDHM Renda 2000']
                     == df2.where(df2['UF']=='PE')
                     ['IDHM Renda 2000'].min()]
q4
```

```python
#Essa próxima questão, é como se "pulássemos" as etapas da Q1 e da Q3
#05) Qual foi o “IDHM Longevidade 2010” da cidade da Bahia que teve o maior “IDHM Educacao 2000”?

df2['IDHM Longevidade 2010'].loc[df2['IDHM Educacao 2000']
                                 == df2.where(df2['UF']=='BA')
                                 ['IDHM Educacao 2000'].max()]
```

```python
#Porque apareceu tantas respostas?

df2.loc[df2['IDHM Educacao 2000']
        == df2.where(df2['UF']=='BA')
        ['IDHM Educacao 2000'].max()]
```

```python
#Mais uma vez, vamos usar um "filtro de linhas"

df_BA=df2.where(df2['UF']=='BA').dropna()
df_BA.head()
```

```python
#Agora que só temos cidades da BA

q5=df_BA['IDHM Longevidade 2010'].loc[df_BA['IDHM Educacao 2000']
                                      == df2.where(df2['UF']=='BA')
                                      ['IDHM Educacao 2000'].max()]
q5
```
