```python
#Preparando o DF:

df['Taxa de crescimento']=df['Taxa de crescimento'].str.strip('%').astype(float)

df['Cidade']=df['Município'].str.split('(',expand=True)[0]
df['UF']=df['Município'].str.split('(',expand=True)[1].str.strip(')')

df2=df.filter(items=['UF','Cidade', 'População em 2010',
                     'População em 2022', 'Variação absoluta',
                     'Taxa de crescimento'])

df2.head()
```

* Q01)  Encontre qual a cidade:
>* de SC que teve a maior Taxa de Crescimento
>* do PA que teve a menor População em 2022
>* quantas cidades (e quais) do RS tiveram a População em 2022, abaixo da média da População em 2022 do PI

```python
q1_1 = df2['Cidade'].loc[df2['Taxa de crescimento']
                       == df2.where(df2['UF']=='SC')
                       ['Taxa de crescimento'].max()]
q1_1
```

```python
q1_2=df2['Cidade'].loc[df2['População em 2022']
                        == df2.where(df2['UF']=='PA')
                        ['População em 2022'].min()]
q1_2
```

Nessa aqui temos que testar algumas coisas primeiro...

```python
df2['Cidade'].loc[df2['População em 2022']
                        < df2.where(df2['UF']=='PI')
                        ['População em 2022'].mean()]
```

```python
df2['UF'].loc[df2['População em 2022']
                        < df2.where(df2['UF']=='PI')
                        ['População em 2022'].mean()]
```

```python
#Posso usar esse mesmo código para contar quantas cidades do RS: "value_counts()"

df2['UF'].loc[df2['População em 2022']
                        < df2.where(df2['UF']=='PI')
                        ['População em 2022'].mean()].value_counts()
```

**Precisamos usar "dois filtros" em duas abordagens diferentes:**

* 1o para encontrar quais cidades (linhas) têm essa característica
* 2o para encontrar quais delas são do RS

**... ou ao contrário:**

* 1o para encontrar quais delas são do RS
* 2o para encontrar quais cidades têm essa característica

```python
#Para encontrar mais "rápido", NÃO vou especificar a coluna do DF2

q1_3a = df2.loc[df2['População em 2022']
                        < df2.where(df2['UF']=='PI')
                        ['População em 2022'].mean()]
q1_3a
```

```python
q1_3b = q1_3a.where(q1_3a['UF']=='RS')['Cidade'].dropna()
q1_3b
```

**Testando a 2a abordagem**

```python
df_RS = df2.where(df2['UF']=='RS').dropna()
df_RS
```

```python
q1_3c = df_RS['Cidade'].loc[df_RS['População em 2022']
                        < df2.where(df2['UF']=='PI')
                        ['População em 2022'].mean()]
q1_3c
```

* Q02) Encontre qual a População em 2010:
>* dessa cidade de SC
>* dessa cidade do PA
>* dessas cidades do MS

```python
q2_1 = df2['População em 2010'].loc[df2['Taxa de crescimento']
                       == df2.where(df2['UF']=='SC')
                       ['Taxa de crescimento'].max()]
q2_1
```

```python
q2_2 = df2['População em 2010'].loc[df2['População em 2022']
                        ==df2.where(df2['UF']=='PA')
                        ['População em 2022'].min()]
q2_2
```

```python
#Nesse caso, vou usar o trecho "where+loc" em todo o DF2 e não apenas em uma coluna específica

q2_3a = df2.loc[df2['População em 2022']
                        < df2.where(df2['UF']=='PI')
                        ['População em 2022'].mean()]
```

```python
#Esse aqui é um "mini-dataframe" que foi "recortado" do DF2 com os critérios do enunciado

q2_3a
```

```python
q2_3b = q2_3a.where(q2_3a['UF']=='RS')['População em 2010'].dropna()
q2_3b
```

```python
#Poderia escrever "tudo junto", mas fica muito difícil de ler e de entender:

df2.loc[df2['População em 2022']<df2.where(df2['UF']=='PI')['População em 2022'].mean()].where(df2['UF']=='RS')['População em 2010'].dropna()
```
