## **Preparando o DF**

```python
import pandas as pd

df=pd.read_csv('https://github.com/marcelovc785/AP2_PML_8001/raw/main/IDSC_BR_AP2_E01d2G01.csv', encoding='ISO-8859-1', sep=';', decimal=',')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
df.Territorialidade.str.split(';',expand=True)
```

```python
df['Cidade']=df.Territorialidade.str.split(';',expand=True)[1]
df['UF']=df.Territorialidade.str.split(';',expand=True)[2]
df['Região']=df.Territorialidade.str.split(';',expand=True)[3]
df.head()
```

```python
#Parâmetros do simulado
#ano1: 2019
#ano2: 2022
#coluna1: ODS 1
#coluna2: ODS 3
#coluna3: ODS 5
```

Como a coluna '2022 ODS 5' e a coluna '2019 ODS 5' estão IGUAIS, vou trocar o ENUNCIADO e vamos considerar a coluna do ODS 3 para verificar essa taxa de crescimento

```python
df['taxa_col2']=(df['2022 ODS 3'] - df['2019 ODS 3'])/df['2019 ODS 3']
df.head()
```

```python
df.columns
```

```python
df2=df.filter(items=['Região', 'UF', 'Cidade', 'População_Ref',
                     'Pontuação Indice ODS 2022','Classificação 2022',
                     '2022 ODS 1', '2022 ODS 3', '2022 ODS 5',
                     'Pontuação Indice ODS 2019', 'Classificação 2019',
                     '2019 ODS 1', '2019 ODS 3', '2019 ODS 5',
                     'taxa_col3', 'taxa_col2'])
df2.head()
```
