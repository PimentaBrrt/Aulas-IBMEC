```python
# Entendendo o enunciado
def notas_final(turma):
  print("Carla", turma["Carla"][0])
  print(sum(turma["Carla"]) / 3)
  print("José", turma["José"][0])
  print(sum(turma["José"]) / 3)
  print("Camila", turma["Camila"][0])
  print(sum(turma["Camila"]) / 3)
  soma_carla = 0
  soma_jose = 0
  soma_camila = 0
  for i in range(3):
    soma_carla += turma["Carla"][i]
    soma_jose += turma["José"][i]
    soma_camila += turma["Camila"][i]
  return (soma_carla + soma_jose + soma_camila) / 9

turma01 = {"Carla": [1,2,3], "José": [0,6,9], "Camila": [4,7.5,10]}
media_turma = notas_final(turma01)
print(media_turma)
```

```python
# Resolução final
carla = (4.2 + 9 + 5.4) / 3
jose = (5.2 + 7.3 + 7) / 3
camila = 6.5
print(carla)
print(jose)
print(camila)
media = (carla + jose + camila) / 3
print(media) # -> Arrendoda para 6.4
```
